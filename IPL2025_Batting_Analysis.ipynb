```python
import pandas as pd  
import numpy as np 
import matplotlib.pyplot as plt  
import seaborn as sns  
```

```python
data = pd.read_csv(r"C:\Users\HP\Desktop\Data Analysis\New folder\IPL 2025 Analysis\IPL2025Batters.csv")
data.info()
```

```python
data['HS'] = data['HS'].str.replace('*','')
data.head()
```

```python
data[['HS','AVG']] = data[['HS','AVG']].apply(pd.to_numeric,errors='coerce')
data.info()
```

```python
# Top 10 Scorers 

data.head(10)
```

```python
# Players with highest batting average 
data.sort_values('AVG',ascending=False).head(10)
```

```python
data[['Player Name','AVG']].sort_values('AVG',ascending=False).head(10)
```

```python
# Players with Highest strike rate 

data.sort_values('SR',ascending=False).head(10)
```

```python
data[['Player Name','SR']].sort_values('SR',ascending=False).head(10)
```

```python
# Most centuries (100s)

data.sort_values('100s',ascending=False).head()
```

```python
# Most half-centuries (50s)

data[['Player Name','50s']].sort_values('50s',ascending=False).head()
```

```python
# Highest individual score (HS)

data[['Player Name','HS']].sort_values('HS',ascending=False).head()
```

```python
# Total runs scored by each team

gb = data.groupby('Team')['Runs'].sum().sort_values(ascending=False).head()
print(f"Total runs Scored by each team {gb}")
```

```python
# Average strike rate per team
gb1 = data.groupby('Team')['SR'].mean().sort_values(ascending=False).head()
print(f"Average Strike Rate per team {gb1}")
```

```python
# Which team hit the most sixes?

gb3 = data.groupby('Team')['6s'].sum().sort_values(ascending=False).head()
print(f"Teams hit the most sixes {gb3}")
```

```python
# Which team has the most 50+ scores?

gb3 = data.groupby('Team')['50s'].sum().sort_values(ascending=False).head()
print(f"Teams with the most 50s {gb3}")
```

```python
# Most consistent player (high avg + low not outs)
data['Consistency_Score'] = data['AVG'] / (data['No'] + 1)
most_consistent = data[['Player Name', 'AVG', 'No', 'Consistency_Score']].sort_values('Consistency_Score', ascending=False).head(10)
print(most_consistent)
```

```python
# Players with highest boundary-percentage (4s+6s per run)

data['Runs by Boundary'] = (data['4s'] * 4) + (data['6s'] * 6)
data['Boundary Percentage'] = ((data['Runs by Boundary']) / data['Runs']) * 100
boundary_percentage = data[['Player Name','4s','6s','Runs by Boundary','Boundary Percentage']].sort_values('Boundary Percentage',ascending=False).head()
print("Players with the highest boundary-percentage")
print(boundary_percentage)
```

```python
data.head()
```

```python
# Fastest scorers (SR above 160)
fastest_scorers = data[data['SR'] > 160][['Player Name', 'SR']].sort_values('SR', ascending=False)
print("Players with Strike Rate above 160:")
print(fastest_scorers.head(10))
```

```python
# Players with 0 centuries but high runs

players = data[data['100s'] == 0][['Player Name','Runs','100s']].sort_values('Runs',ascending=False).head()
print(players)
```

```python
# Bar plot for top 10 run scores

top_scores = data[['Player Name', 'Runs']].sort_values('Runs', ascending=False).head(10)

plt.figure(figsize=(8,5))
bars = plt.bar(top_scores['Player Name'], top_scores['Runs'], color='skyblue')
plt.xlabel('Player Name')
plt.ylabel('Runs')
plt.title('Top 10 Run Scorers')
plt.xticks(rotation=45)


for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, yval + 5, int(yval), ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

```python
# Pie chart of total 4s and 6s by team

gb4 = data.groupby('Team')[['4s','6s']].sum()
gb4['Total Boundaries'] = gb4['4s'] + gb4['6s']
plt.pie(gb4['Total Boundaries'],labels=gb4.index,autopct='%1.1f%%',startangle=90)
plt.title("Pie Chart of total 4s and 6s by team")
plt.axis('equal')
plt.tight_layout()
plt.show()
```

```python
# Heatmap: Correlation between SR, AVG, 4s, 6s

corr_data = data[['SR','AVG','4s','6s']]

corr_matrix = corr_data.corr()

plt.figure(figsize=(6,5))
sns.heatmap(corr_matrix,annot=True,cmap='YlGnBu', fmt='.2f')
plt.title("Correlation Between SR, AVG, 4s, 6s")
plt.tight_layout()
plt.show()
```

```python
# Line plot: Player performance trend (if time-series is added)

plt.figure(figsize=(12, 6))

# Create the box plot
sns.boxplot(x='Team', y='Runs', data=data)

# Add titles and labels
plt.title('Distribution of Player Runs by Team', fontsize=14)
plt.xlabel('Team', fontsize=12)
plt.ylabel('Runs', fontsize=12)
plt.xticks(rotation=45)
plt.tight_layout()

# Show the plot
plt.show()
```
